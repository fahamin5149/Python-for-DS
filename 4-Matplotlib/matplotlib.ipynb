```python
import numpy as np

x=np.linspace(0,10, 5) # from zero to 10, 5 linearly spaced
y=x**2

x,y
```

```python
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#### Multiple graph

```python
plt.subplot(1, 2, 1) #number of rows, number of columns and the current graph
plt.plot(x,y, 'r')

plt.subplot(1,2, 2) #number of rows, number of columns and the current graph
plt.plot(x,y,'g')

plt.show()
```

Sub plots with Object oriented

```python
fig, axes= plt.subplots(nrows=1, ncols=2) #axes is list of plots
axes[0].plot(x,y)
axes[1].plot(x,y)
```

Figsize in subplots

```python
fig, axes= plt.subplots(nrows=2,ncols=2,figsize=(10,8)) 
axes[0,0].plot(x,y)
```

```python
import matplotlib.pyplot as plt

# Example data
years = [2018, 2019, 2020, 2021, 2022]
sales = [100, 150, 200, 250, 300]  # Sales data
profit = [20, 30, 50, 70, 90]  # Profit data

fig, ax1 = plt.subplots()

# Plot sales vs year on the primary y-axis
ax1.plot(years, sales, 'b-o', label='Sales')

ax1.set_xlabel('Year')
ax1.set_ylabel('Sales', color='b')

# Create a secondary y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(years, profit, 'g-s', label='Profit')
ax2.set_ylabel('Profit', color='g')

plt.title('Sales and Profit vs Year')
fig.tight_layout()  # Adjust layout to prevent overlap
plt.grid(True)

plt.show()
```
