```python
import pandas as pd
import numpy as np
```

```python
from plotly import __version__

print(__version__)
```

```python
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df= pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
df
```

```python
df2= pd.DataFrame({'Categories':['A', "B", 'C'], 'Values': [ 32, 43, 50]})
df2
```

```python
df.iplot()
```

```python
df.iplot( kind= "scatter", x="A", y='B') # this is the default scatter in which the points are connected with line
```

```python
df.iplot( kind= "scatter", x="A", y='B', mode= 'markers') 
```

```python
df2.iplot( kind= "bar", x="Categories", y='Values', mode= 'markers') 
```

```python
df.iplot(kind= "box")
```

```python
df3= pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], '2':[500, 400, 300, 200, 180]})
df3
```

```python
df3.iplot(kind= "surface")
```

```python
df['A'].iplot(kind= 'hist')
```

Spread plot is used when we are comparing two columns for example stock price of two companies

```python
df[['A', 'B']].iplot(kind='spread')
```

This is similar to pairplot of the seaborn library and for this all the columns should be numerical
