```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

### Univariate distribution (distribution of a single column)

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False) # kde is a kernal distribution line which is drawn above
```

```python
sns.distplot(tips["total_bill"],bins=10)
```

### Bivariate distribution (distribution of a single column)

```python
sns.jointplot(x= tips['total_bill'], y= tips['tip'])
```

```python
"""
the kind argument is to edit the center plot which was scatter before 
and now changed to hex which is hexagones instead of scatter plot.
"""
sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='hex') 
```

```python
"""
Another thing is to pass the columns and dataset separately
reg is a regression line
"""
sns.jointplot(x= 'total_bill', y='tip', data= tips, kind='reg')
```

```python
"""
Density where the points reach the most
"""

sns.jointplot(x= 'total_bill', y='tip', data= tips, kind='kde')
```

### Pairplot create jointplot for all the possible column

```python
sns.pairplot(tips)
```

```python
"""
passing the categorical column in hue
"""

sns.pairplot(tips, hue= 'sex')
```

```python
"""
using this we build kde 
"""
sns.rugplot(tips['total_bill'])
```

### Barplot

```python
"""
This will take the mean of the male and female data
"""

sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
"""
For any function other than mean, use estimator 
"""
import numpy as np

sns.barplot(x='sex', y= 'total_bill', data=tips, estimator= np.std)
```

### Count plot 

Only x cordinate is passed and y cordinate will automatically be the count of that category

```python
sns.countplot(x= tips['sex'])
```
