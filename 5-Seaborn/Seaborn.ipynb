```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

### Univariate distribution (distribution of a single column)

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False) # kde is a kernal distribution line which is drawn above
```

```python
sns.distplot(tips["total_bill"],bins=10)
```

### Bivariate distribution (distribution of a single column)

```python
sns.jointplot(x= tips['total_bill'], y= tips['tip'])
```

```python
"""
the kind argument is to edit the center plot which was scatter before 
and now changed to hex which is hexagones instead of scatter plot.
"""
sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='hex') 
```

```python
"""
Another thing is to pass the columns and dataset separately
reg is a regression line
"""
sns.jointplot(x= 'total_bill', y='tip', data= tips, kind='reg')
```

```python
"""
Density where the points reach the most
"""

sns.jointplot(x= 'total_bill', y='tip', data= tips, kind='kde')
```

### Pairplot create jointplot for all the possible column

```python
sns.pairplot(tips)
```

```python
"""
passing the categorical column in hue
"""

sns.pairplot(tips, hue= 'sex')
```

```python
"""
using this we build kde 
"""
sns.rugplot(tips['total_bill'])
```

### Barplot

```python
"""
This will take the mean of the male and female data
"""

sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
"""
For any function other than mean, use estimator 
"""
import numpy as np

sns.barplot(x='sex', y= 'total_bill', data=tips, estimator= np.std)
```

### Count plot 

Only x cordinate is passed and y cordinate will automatically be the count of that category

```python
sns.countplot(x= tips['sex'])
```

### Box plot

```python
"""
The data outside the plots are the outliers
"""

sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
"""
Now two box plots are drawn for each
One is the day corresponding the bill given that the person is smoker and 
the second is day to bill given the person is not smoker
"""

sns.boxplot(x='day', y= 'total_bill', data=tips, hue='smoker')
```

### Voilenplot

```python
"""
This tells the distribution of the data, where most of the data points are
"""

sns.violinplot(x='day', y= 'total_bill', data=tips)
```

### Strip plot 

A scatter plot for the categorical data

```python
sns.stripplot(x='day', y='total_bill', data = tips)
```

```python
sns.stripplot(x='day', y='total_bill', data = tips, hue= 'sex', jitter=True)
```

### Swarm plot 

Combination of strip plot and violen plot 

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

### Matrix plot

```python
flights= sns.load_dataset('flights')
flights
```

### Heat map

This require the data in the format of some matrix where there is some relation between rows and columns

```python
tx=tips.corr()
tx
```

```python
sns.heatmap(tx)
```

```python
sns.heatmap(tx, annot=True)
```

```python
sns.heatmap(tx, annot =True, cmap='coolwarm')
```

```python
flights.head()
```

```python
"""
Now converting the flights df to a matrix
In the below pivot table, the index which is x-axis is month and the 
y-axis is year and the value would be passenger
"""

fp=flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
"""
Now as we move forward in years the number of flights increase
"""

sns.heatmap(fp)
```

```python
"""
Magma: Darker to light
"""
sns.heatmap(fp, cmap='magma', linecolor='black', linewidths=2)
```

### Cluster map

Most similar are clustered together

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

### Grids

```python
iris= sns.load_dataset('iris')
```

```python
iris
```

```python
sns.pairplot(iris)
```

### PairGrid

Now we have more control over the plots

```python
g= sns.PairGrid(iris)
```

```python
"""
PairGrid is similar to pairplot, but in pairgrid we have more control to the plots

"""

g= sns.PairGrid(iris)
g.map_diag(sns.boxplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

### FacetGrid

All the possible combination of row and column and create plot for example displot of total bill when there is smoker=yes and time=Lunch

```python
g= sns.FacetGrid(data=tips, row='smoker', col='time')
g.map(sns.distplot, 'total_bill')
```

```python
"""
We are creating a scatter plot in which x axis is total bill and y axis is tip, but we are using two categorical columns
as conditions smoker and time and it will use all the possible combination of them.
"""

g= sns.FacetGrid(data=tips, row='smoker', col='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
```
